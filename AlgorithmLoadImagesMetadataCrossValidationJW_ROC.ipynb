## Load Image Data and Resize

```python
from skimage import io, transform 
from os import listdir, path, makedirs
from os.path import isfile, join
import numpy as np
import pandas as pd

# https://stackoverflow.com/questions/3207219/how-do-i-list-all-files-of-a-directory
raw_image_path = '/Users/JessicaSibal/archive'
raw_image_files = [f for f in listdir(raw_image_path) 
                        if not f.startswith(".")
                            if isfile(join(raw_image_path, f))]

# Issues with Mac hidden files, function to remove these files
#def listdir_nohidden(raw_image_path):
#    for f in listdir(raw_image_path):
#        if not f.startswith('.'):
#            yield f
        
#raw_image_files = listdir_nohidden(raw_image_path)

# Resize Path
resized_image_path = '/Users/JessicaSibal/ResizedImages2'
if not path.exists(resized_image_path):
    makedirs(resized_image_path)

            
# I arbitrarily chose the size to be 100x300 here
# this will fail for gray scale images (but hopefully none are uploaded?)
new_shape = [100,300,3]
for f in raw_image_files:
    # get file names
    filename_read = join(raw_image_path, f)
    filename_write = join(resized_image_path, f)
    
    # read in
    print('Reading ',filename_read)
    # normalize and make float (better for resized interpolation)
    img = io.imread(filename_read)/255.0
    img = img[:,:,:3]
    img = img.astype(np.float)
    print('of size, Resizing...', img.shape)
    
    # resize and save
    img = transform.resize(img,new_shape)
    io.imsave(arr=img,fname=filename_write,mode='constant')
    print('Success!')
    print(len(raw_image_files))
```

## Load Metadata, Add State Label Column

```python
import pandas as pd

df=pd.read_csv('/Users/JessicaSibal/bipolar_data.csv') #CHANGE THIS TO YOUR PATH

# define function to set state codes into labels
def state_label(state):
    if state == 0:
        return 'manic'
    elif state == 1:
        return 'depressed'
    elif state == 2:
        return 'stable'
    else:
        return 'other'

# Adding column for manic, depressed, and stable
df['state_label'] = df['state'].apply(state_label)

len(df)
```

## Add Column for Euphoric, Dysphoric and Psychosis

```python
# creating three new columns that indicate Euphoric, Dysphoric and Psychosis based on the different feeling fields 
#     from the survey. Can combine these into one column but need to figure out what to do in cases when 
#     feelings are a combination of Euphoric, Dysphoric or Psychosis.

# create indicator to show euphoric
df['IsEuphoric'] = (df['feeling_0'] == 1) | (df['feeling_3'] == 1) | (df['feeling_4'] == 1) | (df['feeling_7'] == 1) | (df['feeling_9'] == 1) | (df['feeling_12'] == 1)  
df.IsEuphoric = df.IsEuphoric.astype(np.int)

# create indicator to show dysphoric
df['IsDysphoric'] = (df['feeling_1'] == 1) | (df['feeling_2'] == 1) | (df['feeling_5'] == 1) | (df['feeling_6'] == 1) | (df['feeling_8'] == 1) | (df['feeling_10'] == 1)  
df.IsDysphoric = df.IsDysphoric.astype(np.int)

# Create psychosis indicator
df['IsPsychosis'] = (df['feeling_11'] == 1) | (df['feeling_13'] == 1)  
df.IsPsychosis = df.IsPsychosis.astype(np.int)

df.head()
```

```python
import keras
from keras.models import Sequential
from keras.layers import Reshape
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.layers import Conv2D, MaxPooling2D
from keras.callbacks import EarlyStopping
from keras.regularizers import l2 
import numpy as np
l2_lambda = 0.0001

keras.__version__
```

```python
# prepare labels for use in keras 
from sklearn.preprocessing import LabelEncoder

# convert string to integers
enc = LabelEncoder()
y_as_ints = enc.fit_transform(labels)

# convert to one hot encoded vector (this is optional depending on how you create the model)
NUM_CLASSES = len(np.unique(labels))
y_ohe = keras.utils.to_categorical(y_as_ints, NUM_CLASSES)

X = img_features # this is our training data
```

```python
# create an architecture, with data expansion

from keras.preprocessing.image import ImageDataGenerator

# create a mechanism for randomly perturbing the image data
# after each epoch (this is for better generalization)
datagen = ImageDataGenerator(featurewise_center=False,
    samplewise_center=False,
    featurewise_std_normalization=False,
    samplewise_std_normalization=False,
    zca_whitening=False,
    rotation_range=5, # used, Int. Degree range for random rotations.
    width_shift_range=0.1, # used, Float (fraction of total width). Range for random horizontal shifts.
    height_shift_range=0.1, # used,  Float (fraction of total height). Range for random vertical shifts.
    shear_range=0., # Float. Shear Intensity (Shear angle in counter-clockwise direction as radians)
    zoom_range=0.,
    channel_shift_range=0.,
    fill_mode='nearest',
    cval=0.,
    horizontal_flip=True,
    vertical_flip=False,
    rescale=None)

datagen.fit(X)
```

```python
# create the architecture to use for the model
# create a method to re-initialze model so that we can do k-fold with fresh new model at each fold
def model():
    cnn = Sequential()


    # add one conv layer with relu nonlinearity
    cnn.add(Conv2D(filters=32,
                    input_shape = X.shape[1:], # tell it the size of images, rows x cols x channels
                    kernel_size=(3,3), 
                    padding='same', 
                    kernel_initializer='he_uniform', 
                    kernel_regularizer=l2(l2_lambda),
                    activation='relu')) # add relu layer after

    # add in one max pool layer
    cnn.add(MaxPooling2D(pool_size=(2, 2), data_format="channels_last"))

    # typically you would repeat the above layer combinations a few times

    # add one layer on flattened output
    cnn.add(Dropout(0.25)) # add some dropout for regularization after conv layers
    cnn.add(Flatten())

    cnn.add(Dense(128,
                  kernel_initializer='he_uniform',
                  kernel_regularizer=l2(l2_lambda),
                  activation='relu'))

    cnn.add(Dropout(0.5)) # add some dropout for regularization, again!

    cnn.add(Dense(NUM_CLASSES, 
                  activation='softmax'))

    # Let's train the model 
    cnn.compile(loss='categorical_crossentropy', # 'categorical_crossentropy' 'mean_squared_error'
                  optimizer='adam', # 'adadelta' 'rmsprop'
                  metrics=['accuracy'])

    num_runs = 0
    return cnn
```

## Based on the split in feelings for manic our results are not that great

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import StratifiedKFold
kf = KFold(n_splits=10, shuffle=True,random_state=1)
#kf = StratifiedKFold(n_splits=10)
#kf.get_n_splits(X)

y_hat = []

for train_index, test_index in kf.split(X,y_ohe):
    cnn=model() #create a fresh cnn model for each fold
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y_ohe[train_index], y_ohe[test_index]
   
    cnn.fit_generator(datagen.flow(X_train, y_train, batch_size=16), 
                   steps_per_epoch=int(len(X)/16), # how many generators to go through per epoch
                   epochs=5, verbose=1)
    y_hat.append(cnn.predict(X_test, verbose=1))
y_hat
    
# the flow method yields batches of images indefinitely, with the given transformations
num_runs += 1 # this block can be run multiple times and each time gets you 
```

```python
# create an architecture, with data expansion

from keras.preprocessing.image import ImageDataGenerator

# create a mechanism for randomly perturbing the image data
# after each epoch (this is for better generalization)
datagen = ImageDataGenerator(featurewise_center=False,
    samplewise_center=False,
    featurewise_std_normalization=False,
    samplewise_std_normalization=False,
    zca_whitening=False,
    rotation_range=5, # used, Int. Degree range for random rotations.
    width_shift_range=0.1, # used, Float (fraction of total width). Range for random horizontal shifts.
    height_shift_range=0.1, # used,  Float (fraction of total height). Range for random vertical shifts.
    shear_range=0., # Float. Shear Intensity (Shear angle in counter-clockwise direction as radians)
    zoom_range=0.,
    channel_shift_range=0.,
    fill_mode='nearest',
    cval=0.,
    horizontal_flip=True,
    vertical_flip=False,
    rescale=None)

datagen.fit(X)
```

```python
X.shape
```

```python
# create the architecture to use for the model
# create a method to re-initialze model so that we can do k-fold with fresh new model at each fold
def model():
    cnn = Sequential()


    # add one conv layer with relu nonlinearity
    cnn.add(Conv2D(filters=32,
                    input_shape = X.shape[1:], # tell it the size of images, rows x cols x channels
                    kernel_size=(3,3), 
                    padding='same', 
                    kernel_initializer='he_uniform', 
                    kernel_regularizer=l2(l2_lambda),
                    activation='relu')) # add relu layer after

    # add in one max pool layer
    cnn.add(MaxPooling2D(pool_size=(2, 2), data_format="channels_last"))

    # typically you would repeat the above layer combinations a few times

    # add one layer on flattened output
    cnn.add(Dropout(0.25)) # add some dropout for regularization after conv layers
    cnn.add(Flatten())

    cnn.add(Dense(128,
                  kernel_initializer='he_uniform',
                  kernel_regularizer=l2(l2_lambda),
                  activation='relu'))

    cnn.add(Dropout(0.5)) # add some dropout for regularization, again!

    cnn.add(Dense(NUM_CLASSES, 
                  activation='softmax'))

    # Let's train the model 
    cnn.compile(loss='categorical_crossentropy', # 'categorical_crossentropy' 'mean_squared_error'
                  optimizer='adam', # 'adadelta' 'rmsprop'
                  metrics=['accuracy'])

    num_runs = 0
    return cnn
```

```python
cnn=model()
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import StratifiedKFold
kf = KFold(n_splits=10, shuffle=True,random_state=1)
#kf = StratifiedKFold(n_splits=10)
#kf.get_n_splits(X)

y_hat = []

for train_index, test_index in kf.split(X,y_ohe):
    cnn=model() #create a fresh cnn model for each fold
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y_ohe[train_index], y_ohe[test_index]
    #print('Training runs:',num_runs)
    cnn.fit_generator(datagen.flow(X_train, y_train, batch_size=16), 
                   steps_per_epoch=int(len(X)/16), # how many generators to go through per epoch
                   epochs=5, verbose=1)
    y_hat.append(cnn.predict(X_test, verbose=1))
y_hat
    
# the flow method yields batches of images indefinitely, with the given transformations
num_runs += 1 # this block can be run multiple times and each time gets you 

              
              
```

```python
def per_class_accuracy(ytrue,yhat):
    conf = mt.confusion_matrix(ytrue,yhat)
    norm_conf = conf.astype('float') / conf.sum(axis=1)[:, np.newaxis]
    return np.diag(norm_conf)

def plot_class_acc(ytrue,yhat, title=''):
    acc_list = per_class_accuracy(ytrue,yhat)
    plt.bar(range(len(acc_list)), acc_list)
    plt.xlabel('Class value (one per face)')
    plt.ylabel('Accuracy within class')
    plt.title(title+", Total Acc=%.1f"%(100*mt.accuracy_score(ytrue,yhat)))
    plt.grid()
    plt.ylim([0,1])
    plt.show()
```
